## Missing Values

> Como podemos observar pelo excerto abaixo, existem 1440 rows em que não existem informações sobre o setor e a indústria da empresa

> Podiamos completar estes valores com os setores/indústrias mais vistos segundo este dataset, mas estaríamos a adulterar os dados, tendo em conta que estes campos têem imensa variadade.

> Decidimos então excluir todas as empresas que não tenham estes valores, guardando-os num array para que posteriormente sejam eliminados dos stock prices.

> Por norma, todas as rows que não contém o setor não contém a indústria e vice-versa

```python
stock_companies.isnull().sum()
```

```python
null_data = stock_companies[stock_companies.isnull().any(axis=1)]
bad_tickers = null_data['ticker'].unique()
null_data
```

> Removemos todos os missing values

```python
stock_companies = stock_companies.dropna(how='any',axis=0) 
stock_companies.isnull().sum()
```

### Após termos a dataset preparado decide-se fazer uma pequena exploração, às empresas presentes, relativamente:

1. Qual o setor predominante.
2. Quantidade de companhias petencentes a cada stock exchange

```python
def make_plots(df,label):
    
    x = stock_companies[label].unique()
    
    x_size = len(x)

    tmp = np.zeros(shape=(x_size,))
    
    for i in range(x_size):
        n = x[i]
        tmp[i] = len(df[df[label] == n])

    raw_data = { label : x,
                'count' : tmp
    }

    df = pd.DataFrame(raw_data, columns = [label, 'count'])
    ax = df.plot.barh(x=label, y='count')
    plt.show()
```

> Como podemos observar os 3 setores predominantes são, health care(o que é inesperado), consumer services e Finance, sendo que o setor de Finanças é o mais esperado neste caso.

```python
make_plots(stock_companies,'sector')
```

> Apenas temos dois stocks exchange, depois explicar..

```python
make_plots(stock_companies,'exchange')
```
